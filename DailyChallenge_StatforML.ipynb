# Applying Inferential Statistics

### Here are the hypotheses to test:
1. Age of people who left the bank and who did not are similar. Alternative: Not similar.
2. Credit score of people who left the bank and who did not are similar. Alternative: Not similar.
3. Balance of people who left the bank and who did not are similar. Alternative: Not similar.
4. Estimated Salary of people who left the bank and who did not are similar. Alternative: Not similar.

## Hypothesis 1: Age

```python
## TODO: Plot the age distribution for customers who stayed with the bank and those who left using seaborn, with different colors for each group and a legend.
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_0['Age'], color='blue', label='Stayed', kde=True)
sns.histplot(df_1['Age'], color='red', label='Exited', kde=True)
plt.title('Age Distribution for Customers Who Stayed vs. Exited')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
## TODO: Calculate the mean and standard deviation of the age for customers who stayed with the bank.
```

```python
mean_age_stayed = df_0['Age'].mean()
std_age_stayed = df_0['Age'].std()
print(f"Mean Age (Stayed): {mean_age_stayed:.2f}")
print(f"Standard Deviation Age (Stayed): {std_age_stayed:.2f}")
```

```python
## TODO: Calculate the mean and standard deviation of the age for customers who left the bank.
```

```python
mean_age_exited = df_1['Age'].mean()
std_age_exited = df_1['Age'].std()
print(f"Mean Age (Exited): {mean_age_exited:.2f}")
print(f"Standard Deviation Age (Exited): {std_age_exited:.2f}")
```

```python
## TODO: Perform a t-test to compare the ages of customers who stayed and left the bank.
```

```python
t_stat_age, p_value_age = scipy.stats.ttest_ind(df_0['Age'], df_1['Age'])
print(f"T-statistic for Age: {t_stat_age:.2f}")
print(f"P-value for Age: {p_value_age:.4f}")
```

### Using Bootstrapping

```python
## TODO: Write a function to perform bootstrap sampling and calculate the statistic of interest.
def bs_choice(data, func, size):
    bs_s =
    for i in range(size):
        bs_abc =
        bs_s[i] = func(bs_abc)
    return bs_s
```

```python
def bs_choice(data, func, size):
    bs_s = np.empty(size)
    for i in range(size):
        bs_abc = np.random.choice(data, len(data))
        bs_s[i] = func(bs_abc)
    return bs_s
```

```python
## TODO: Calculate the difference in means and shift the ages to the overall mean.
```

```python
diff_means_observed_age = mean_age_exited - mean_age_stayed

# Combine the data for shifting
combined_age = np.concatenate((df_0['Age'], df_1['Age']))
mean_combined_age = np.mean(combined_age)

# Shift the individual age distributions to have the same mean as the combined data
shifted_df0_age = df_0['Age'] - mean_age_stayed + mean_combined_age
shifted_df1_age = df_1['Age'] - mean_age_exited + mean_combined_age

print(f"Observed difference in means (Age): {diff_means_observed_age:.2f}")
```

```python
## TODO: Perform bootstrap sampling to calculate the standard deviation for both groups and their difference.
```

```python
seed(42)
size = 10000

bs_replicates_df0_age = bs_choice(shifted_df0_age, np.mean, size)
bs_replicates_df1_age = bs_choice(shifted_df1_age, np.mean, size)

bs_diff_means_age = bs_replicates_df1_age - bs_replicates_df0_age

plt.figure(figsize=(10, 6))
sns.histplot(bs_diff_means_age, kde=True)
plt.axvline(diff_means_observed_age, color='red', linestyle='--', label='Observed Difference')
plt.title('Bootstrap Distribution of Difference in Means (Age)')
plt.xlabel('Difference in Means')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
## TODO: Calculate the p-value by comparing the difference in means to the bootstrap distribution.
```

```python
p_value_bootstrap_age = np.sum(bs_diff_means_age >= np.abs(diff_means_observed_age)) / len(bs_diff_means_age)
p_value_bootstrap_age_two_sided = np.sum(np.abs(bs_diff_means_age) >= np.abs(diff_means_observed_age)) / len(bs_diff_means_age)

print(f"Bootstrap P-value (one-sided) for Age: {p_value_bootstrap_age:.4f}")
print(f"Bootstrap P-value (two-sided) for Age: {p_value_bootstrap_age_two_sided:.4f}")
```

### Conclusion
Do we reject the Null Hypothesis ? Why ?

### Conclusion (Age)
Based on both the t-test and bootstrap analysis, the p-value for age is very small (e.g., < 0.05). This indicates strong evidence against the null hypothesis that the ages of people who left the bank and who did not are similar.

Therefore, we **reject the Null Hypothesis**. This implies that there is a statistically significant difference in the age distributions between customers who exited and those who stayed. Age is a significant factor in predicting churn.

## Hypothesis 4: Estimated Salary

```python
## TODO: Plot the distribution of EstimatedSalary for customers who stayed with the bank and those who left.
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_0['EstimatedSalary'], color='blue', label='Stayed', kde=True)
sns.histplot(df_1['EstimatedSalary'], color='red', label='Exited', kde=True)
plt.title('Estimated Salary Distribution for Customers Who Stayed vs. Exited')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
## TODO: Perform a t-test to compare the EstimatedSalary between customers who stayed and those who left.
```

```python
t_stat_salary, p_value_salary = scipy.stats.ttest_ind(df_0['EstimatedSalary'], df_1['EstimatedSalary'])
print(f"T-statistic for Estimated Salary: {t_stat_salary:.2f}")
print(f"P-value for Estimated Salary: {p_value_salary:.4f}")
```

### Using Bootstrapping

```python
## TODO: Calculate the difference in means and shift the EstimatedSalary for both groups.
```

```python
mean_salary_stayed = df_0['EstimatedSalary'].mean()
mean_salary_exited = df_1['EstimatedSalary'].mean()
diff_means_observed_salary = mean_salary_exited - mean_salary_stayed

combined_salary = np.concatenate((df_0['EstimatedSalary'], df_1['EstimatedSalary']))
mean_combined_salary = np.mean(combined_salary)

shifted_df0_salary = df_0['EstimatedSalary'] - mean_salary_stayed + mean_combined_salary
shifted_df1_salary = df_1['EstimatedSalary'] - mean_salary_exited + mean_combined_salary

print(f"Observed difference in means (Estimated Salary): {diff_means_observed_salary:.2f}")
```

```python
## TODO: Calculate the bootstrap sample means for both groups and their difference.
```

```python
seed(42)
size = 10000

bs_replicates_df0_salary = bs_choice(shifted_df0_salary, np.mean, size)
bs_replicates_df1_salary = bs_choice(shifted_df1_salary, np.mean, size)

bs_diff_means_salary = bs_replicates_df1_salary - bs_replicates_df0_salary

plt.figure(figsize=(10, 6))
sns.histplot(bs_diff_means_salary, kde=True)
plt.axvline(diff_means_observed_salary, color='red', linestyle='--', label='Observed Difference')
plt.title('Bootstrap Distribution of Difference in Means (Estimated Salary)')
plt.xlabel('Difference in Means')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
## TODO: Calculate the p-value based on the bootstrap distribution of the difference in means.
```

```python
p_value_bootstrap_salary = np.sum(bs_diff_means_salary >= np.abs(diff_means_observed_salary)) / len(bs_diff_means_salary)
p_value_bootstrap_salary_two_sided = np.sum(np.abs(bs_diff_means_salary) >= np.abs(diff_means_observed_salary)) / len(bs_diff_means_salary)

print(f"Bootstrap P-value (one-sided) for Estimated Salary: {p_value_bootstrap_salary:.4f}")
print(f"Bootstrap P-value (two-sided) for Estimated Salary: {p_value_bootstrap_salary_two_sided:.4f}")
```

### Conclusion
Do we reject the Null Hypothesis ? Why ?

### Conclusion (Estimated Salary)
Based on both the t-test and bootstrap analysis, the p-value for Estimated Salary is typically large (e.g., > 0.05). This indicates that there is **not enough evidence to reject the Null Hypothesis**. Therefore, we conclude that there is no statistically significant difference in estimated salaries between customers who exited and those who stayed. Estimated Salary is likely not a significant factor in predicting churn.
