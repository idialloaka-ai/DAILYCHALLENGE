# Défi quotidien : Incitations précises pour un contrôle et une qualité de production optimaux

Ce notebook contient le flux de travail pour transformer une politique technique dense en contenu de micro-apprentissage clair en utilisant des techniques d'incitation avancées.

```python
# Texte d'entrée original (La source de vérité)
policy_text = """
Les employés doivent s'assurer que tout accès distant aux systèmes internes est établi via le VPN sécurisé approuvé. En aucun cas, des connexions non sécurisées ou des appareils personnels dépourvus de protection des terminaux ne doivent être utilisés pour accéder à des données confidentielles ou à des communications sensibles.
"""
print("Texte source chargé.")
```

## Étape 1 & 2 : Création et Évaluation de l'Invite

**Objectif :** Ton amical, puces, paraphrase, < 75 mots.

```python
prompt_v1 = f"""
Tu es un expert en communication interne.
Réécris le texte suivant pour une newsletter d'entreprise :

Texte : {policy_text}

Contraintes :
1. Utilise un ton amical et clair.
2. Organise les informations sous forme de liste à puces.
3. Paraphrase le contenu (ne cite pas directement).
4. Reste strictement en dessous de 75 mots.
"""

# Simulez ici l'envoi à votre modèle d'IA préféré
print("CONSIGNE GÉNÉRÉE :\n", prompt_v1)
```

## Étape 3 : Atténuation des Hallucinations

Si le modèle a ajouté des détails non présents (comme le nom d'un logiciel spécifique), utilisez cette version révisée.

```python
prompt_v2 = prompt_v1 + "\nIMPORTANT : Reste strictement fidèle aux faits fournis. N'ajoute pas de nouvelles technologies ou recommandations non mentionnées dans le texte source."
print("CONSIGNE RÉVISÉE (Anti-hallucination) :\n", prompt_v2)
```

## Étape 4 : Analyse approfondie (Public Junior)

**Objectif :** Langage simple, max 4 puces, ton encourageant.

```python
prompt_junior = f"""
Réécris ce paragraphe de politique pour des stagiaires juniors :

Texte : {policy_text}

Consignes :
- Utilise un langage simple, sans jargon juridique ou technique complexe.
- Adopte un ton encourageant et informatif.
- Utilise au maximum 4 puces.
- Fais des phrases courtes.
"""
print("CONSIGNE POUR STAGIAIRES :\n", prompt_junior)
```

## Étape 5 : Extraction de Citation et Analyse

**Objectif :** Extraire la phrase la plus critique.

```python
prompt_citation = f"""
Extrais la citation directe du texte suivant qui résume le mieux la règle de sécurité essentielle :

Texte : {policy_text}
"""
print("CONSIGNE D'EXTRACTION :\n", prompt_citation)
```

### Analyse de la Citation

**1. Dans quel type de communication interne la citation serait-elle plus appropriée ?**
> Elle est idéale pour des affiches de sensibilisation, des bannières de connexion ou des rappels juridiques officiels où la précision textuelle est obligatoire.

**2. Dans quelles circonstances les citations peuvent-elles présenter un risque ?**
> Les citations peuvent être trop denses ou pleines de jargon, ce qui risque de perdre l'attention de l'employé ou de mener à une mauvaise interprétation si le langage est trop complexe.
