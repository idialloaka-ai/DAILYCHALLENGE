## 1 & 2. Tokenisation du texte
Explorons comment BERT et XLM-RoBERTa décomposent le texte.

```python
from transformers import BertTokenizer, XLMRobertaTokenizer

# Initialisation des tokeniseurs
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
xlmr_tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')

text = "L'apprentissage automatique est fascinant."

# Exemple de tokenisation
bert_input = bert_tokenizer.encode_plus(
    text,
    add_special_tokens=True,
    max_length=20,
    padding='max_length',
    truncation=True,
    return_attention_mask=True,
    return_tensors='pt'
)

print("BERT Tokens ID:", bert_input['input_ids'])
print("BERT Decoded:", bert_tokenizer.decode(bert_input['input_ids'][0]))
print("XLM-R Vocab Size:", xlmr_tokenizer.vocab_size)
```

## 4. Chargement et exploration de l'ensemble de données
Nous allons identifier les fichiers CSV extraits pour charger les données.

```python
import pandas as pd
import os

# Liste des fichiers extraits
files = os.listdir('/content/dataset')
print("Fichiers disponibles :", files)

# Chargement (ajustez les noms si nécessaire selon l'output précédent)
train_df = pd.read_csv('/content/dataset/train.csv') if 'train.csv' in files else pd.DataFrame()

print("Aperçu du dataset :")
display(train_df.head())
print("Forme du dataset :", train_df.shape)
```

## 5. Création de plis de validation croisée (K-Fold)
Utilisation de StratifiedKFold pour garantir une distribution équilibrée des classes.

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Supposons que la colonne cible s'appelle 'label'
if 'label' in train_df.columns:
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    folds_train = []
    folds_val = []

    for train_index, val_index in skf.split(train_df, train_df['label']):
        folds_train.append(train_df.iloc[train_index])
        folds_val.append(train_df.iloc[val_index])

    print(f"Nombre de plis créés : {len(folds_train)}")
    print(f"Taille du premier pli d'entraînement : {len(folds_train[0])}")
else:
    print("La colonne 'label' n'a pas été trouvée. Veuillez vérifier le nom de la colonne cible dans train_df.columns")
```
