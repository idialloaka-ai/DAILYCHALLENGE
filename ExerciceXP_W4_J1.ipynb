## 🌟 Exercice 1 : Définition du problème et collecte de données pour la prédiction des défauts de paiement des prêts

### Énoncé du problème

L'objectif est de développer un modèle d'apprentissage automatique capable de prédire la probabilité qu'un emprunteur fasse défaut sur son prêt. Une détection précoce des emprunteurs à haut risque permettra aux institutions financières de prendre des mesures proactives, telles que l'ajustement des conditions de prêt, l'offre de conseils financiers, ou le refus de prêts, afin de minimiser les pertes financières et de maintenir la stabilité du portefeuille de prêts.

### Types de données nécessaires

Pour prédire les défauts de paiement, les types de données suivants seraient pertinents :

*   **Informations personnelles du demandeur :**
    *   Âge, sexe, état civil, niveau d'éducation, profession.
    *   Revenu annuel, source de revenu (salaire, indépendant, etc.).
    *   Historique de l'emploi (stabilité, durée).
    *   Lieu de résidence (code postal, type de logement).
*   **Informations sur le prêt :**
    *   Montant du prêt demandé, durée du prêt, taux d'intérêt.
    *   Objet du prêt (achat immobilier, voiture, dette, etc.).
    *   Garanties éventuelles.
*   **Scores de crédit et historique financier :**
    *   Score de crédit (par ex. FICO, notation interne).
    *   Historique de paiement (nombre de jours de retard, fréquence des défauts passés).
    *   Nombre de comptes de crédit ouverts, solde total du crédit.
    *   Ratio d'endettement.
    *   Historique des demandes de prêt rejetées.
*   **Données comportementales (si disponibles) :**
    *   Activité bancaire (fréquence des transactions, solde moyen).
    *   Comportement d'épargne.
*   **Variables externes/macroéconomiques :**
    *   Taux de chômage régionaux.
    *   Taux d'inflation.
    *   Taux d'intérêt de référence.

### Sources de données potentielles

*   **Dossiers internes des institutions financières :**
    *   Données historiques sur les clients (informations personnelles, historiques de prêts et de paiements).
    *   Données sur les transactions bancaires et les comptes d'épargne.
*   **Agences d'évaluation du crédit (bureaux de crédit) :**
    *   Scores de crédit (comme Experian, Equifax, TransUnion).
    *   Historiques de crédit détaillés (emprunts passés, habitudes de remboursement, faillites).
*   **Bases de données publiques :**
    *   Statistiques nationales/régionales (données démographiques, taux de chômage de l'INSEE ou d'Eurostat).
    *   Données économiques (Banque Centrale Européenne, organismes gouvernementaux).
*   **Fournisseurs de données tiers :**
    *   Données comportementales ou socio-démographiques enrichies.
    *   Données sur les fraudes ou les risques spécifiques.

##  Exercice 3 : Entraînement, évaluation et optimisation du modèle

### Quel(s) modèle(s) choisiriez-vous pour une prédiction de prêt ?

Compte tenu de la nature sensible de la prédiction des défauts de paiement (où les faux négatifs peuvent entraîner des pertes financières importantes pour la banque), je choisirais un modèle capable de capturer des relations complexes tout en permettant une certaine interprétabilité, et qui gère bien les déséquilibres de classes (le nombre de défauts est généralement bien inférieur au nombre de non-défauts).

Les modèles principaux seraient :

1.  **XGBoost (Extreme Gradient Boosting) :** Très performant et largement utilisé dans l'industrie pour sa précision et sa robustesse. Il gère bien les données déséquilibrées et les valeurs manquantes. Il permet également l'extraction de l'importance des caractéristiques, ce qui est utile pour l'interprétabilité.
2.  **Forêt Aléatoire (Random Forest) :** Un excellent classifieur d'ensemble qui réduit le surapprentissage par rapport aux arbres de décision individuels. Moins complexe à interpréter que XGBoost, mais très efficace.
3.  **Régression Logistique :** Bien que plus simple, elle sert de ligne de base et son interprétabilité est élevée, ce qui est crucial pour comprendre les facteurs de risque et justifier les décisions de prêt.

### Étapes d'évaluation des performances du modèle

L'évaluation d'un modèle de prédiction de défaut de paiement doit être rigoureuse, en se concentrant sur les indicateurs pertinents pour un problème de classification binaire, souvent déséquilibré.

1.  **Division des données :**
    *   Diviser l'ensemble de données en ensembles d'entraînement, de validation (optionnel, pour le réglage des hyperparamètres) et de test (par exemple, 70% entraînement, 15% validation, 15% test). L'ensemble de test doit rester invisible jusqu'à l'évaluation finale.

2.  **Techniques de rééchantillonnage pour le déséquilibre de classes :**
    *   Utiliser des techniques comme SMOTE (Synthetic Minority Over-sampling Technique) sur l'ensemble d'entraînement pour générer des échantillons synthétiques de la classe minoritaire (défauts) ou des méthodes de sous-échantillonnage de la classe majoritaire pour équilibrer la distribution.

3.  **Validation croisée :**
    *   Appliquer la validation croisée (par exemple, K-Fold Cross-Validation) sur l'ensemble d'entraînement/validation pour obtenir une estimation robuste de la performance du modèle et éviter le surapprentissage. Cela permet d'entraîner le modèle plusieurs fois sur différentes partitions des données.

4.  **Métriques d'évaluation spécifiques :**
    *   **Précision (Precision) :** Sur tous les cas prédits comme positifs (défauts), combien étaient réellement des défauts ? `TP / (TP + FP)`. C'est crucial pour la banque pour s'assurer que les ressources ne sont pas gaspillées sur des alertes non fondées.
    *   **Rappel (Recall) / Sensibilité :** Sur tous les cas réels de défauts, combien ont été correctement identifiés ? `TP / (TP + FN)`. C'est extrêmement important pour la banque pour ne pas manquer les vrais cas de défauts, ce qui entraînerait des pertes.
    *   **Score F1 :** La moyenne harmonique de la précision et du rappel. Il est utile lorsque vous avez besoin d'un équilibre entre la précision et le rappel, en particulier avec des classes déséquilibrées.
    *   **Courbe ROC (Receiver Operating Characteristic) et AUC (Area Under the Curve) :** L'AUC mesure la capacité du modèle à distinguer les classes. Une AUC élevée indique que le modèle est bon pour séparer les défauts des non-défauts, indépendamment du seuil de classification choisi. C'est un indicateur robuste pour les problèmes déséquilibrés.
    *   **Matrice de confusion :** Une table qui résume la performance du modèle en montrant les Vrais Positifs (TP), Vrais Négatifs (TN), Faux Positifs (FP) et Faux Négatifs (FN). Elle est essentielle pour comprendre où le modèle se trompe.
    *   **Précision globale (Accuracy) :** `(TP + TN) / (TP + TN + FP + FN)`. Bien que souvent mentionnée, elle peut être trompeuse dans les cas de déséquilibre de classes (un modèle prédisant toujours la classe majoritaire pourrait avoir une précision élevée mais être inutile).

5.  **Optimisation des hyperparamètres :**
    *   Utiliser des techniques comme la recherche par grille (Grid Search) ou la recherche aléatoire (Random Search) combinées à la validation croisée pour trouver les meilleurs hyperparamètres pour le modèle (par exemple, profondeur maximale des arbres, nombre d'estimateurs pour Random Forest/XGBoost, régularisation pour la régression logistique).

6.  **Analyse de l'importance des caractéristiques :**
    *   Pour les modèles basés sur les arbres (Random Forest, XGBoost), analyser l'importance des caractéristiques pour comprendre quels facteurs contribuent le plus à la prédiction. Cela aide à l'interprétabilité et à la validation métier.

7.  **Seuils de décision :**
    *   Ajuster le seuil de décision du classifieur (par défaut 0.5) pour optimiser la précision ou le rappel en fonction des coûts relatifs des faux positifs et des faux négatifs pour l'institution financière.

##  Exercice 5 : Concevoir une stratégie d'évaluation pour différents modèles d'apprentissage automatique

### 1. Modèle d'apprentissage supervisé (Classification) - Exemple : Prédiction de défaut de prêt

#### Stratégie d'évaluation

Pour un modèle de classification (par exemple, prédiction de défaut de prêt) :

1.  **Division des données :** Diviser l'ensemble de données en ensembles d'entraînement (70-80%) et de test (20-30%), en assurant une stratification pour maintenir la proportion des classes dans chaque ensemble, surtout si les classes sont déséquilibrées.
2.  **Validation croisée :** Utiliser la validation croisée (par exemple, 5-fold ou 10-fold Cross-Validation) sur l'ensemble d'entraînement pour entraîner et valider le modèle. Cela permet d'obtenir une estimation plus fiable de la performance en réduisant la variance due à une seule division aléatoire.
3.  **Métriques de performance :**
    *   **Matrice de confusion :** Indispensable pour visualiser les TP, TN, FP, FN.
    *   **Précision (Precision) :** `TP / (TP + FP)`. Utile lorsque le coût des Faux Positifs est élevé.
    *   **Rappel (Recall) / Sensibilité :** `TP / (TP + FN)`. Utile lorsque le coût des Faux Négatifs est élevé (comme dans la détection de défauts).
    *   **Score F1 :** Moyenne harmonique de la précision et du rappel. Un bon indicateur lorsque l'on recherche un équilibre entre les deux, particulièrement avec des classes déséquilibrées.
    *   **Courbe ROC (Receiver Operating Characteristic) et AUC (Area Under the Curve) :** Mesure la capacité du modèle à distinguer les classes pour différents seuils de classification. Une AUC élevée indique une bonne performance générale.
    *   **Courbe Précision-Rappel (PR Curve) :** Souvent plus informative que la courbe ROC pour les problèmes de classification avec des classes très déséquilibrées, car elle se concentre sur la performance de la classe positive.

#### Difficultés et limites

*   **Déséquilibre des classes :** Si la classe de défauts est rare, la précision globale peut être trompeuse. Les métriques comme le rappel et le score F1 deviennent plus importantes.
*   **Coût des erreurs :** Les Faux Positifs (prédire un défaut alors qu'il n'y en a pas) et les Faux Négatifs (ne pas prédire un défaut alors qu'il y en a un) n'ont pas le même coût. L'évaluation doit prendre en compte les implications métier.
*   **Dérive des données (Data Drift) :** Les comportements de défaut peuvent changer avec le temps ou les conditions économiques, rendant les modèles obsolètes. Une réévaluation et un réentraînement périodiques sont nécessaires.

### 2. Modèle d'apprentissage non supervisé (Clustering) - Exemple : Segmentation client

#### Stratégie d'évaluation

Pour un modèle de clustering (par exemple, segmentation client), où il n'y a pas d'étiquettes de vérité terrain, l'évaluation est plus complexe et souvent intrinsèque :

1.  **Visualisation :** Utiliser des techniques de réduction de dimensionnalité (PCA, t-SNE, UMAP) pour visualiser les clusters dans un espace 2D ou 3D et évaluer visuellement leur séparation et leur cohésion.
2.  **Métriques d'évaluation intrinsèques :**
    *   **Score de silhouette :** Mesure la qualité des clusters. Un score élevé indique que les objets sont bien appariés à leur propre cluster et mal appariés aux clusters voisins. La valeur varie de -1 à 1.
    *   **Méthode du coude (Elbow Method) :** Pour déterminer le nombre optimal de clusters en traçant la variance intra-cluster (inertie) en fonction du nombre de clusters. Le 'coude' du graphique suggère le nombre optimal.
    *   **Indice de Davies-Bouldin :** Évalue la similarité moyenne entre chaque cluster et son cluster le plus similaire. Un score faible indique une meilleure séparation des clusters.
    *   **Indice de Calinski-Harabasz (Variance Ratio Criterion) :** Mesure le rapport entre la variance inter-cluster et la variance intra-cluster. Un score plus élevé est préférable.
3.  **Validation métier/domaine :** Examiner les caractéristiques moyennes ou les profils des clusters identifiés et les présenter à des experts du domaine pour vérifier si les groupes ont un sens commercial ou conceptuel. C'est souvent l'évaluation la plus cruciale pour le non-supervisé.

#### Difficultés et limites

*   **Absence de vérité terrain :** La principale difficulté est l'absence d'étiquettes de classe réelles, ce qui rend l'évaluation objective difficile.
*   **Dépendance au prétraitement :** Les résultats du clustering sont très sensibles au prétraitement des données (normalisation, gestion des valeurs aberrantes).
*   **Subjectivité :** L'interprétation des clusters et la détermination du nombre optimal de clusters peuvent être subjectives.
*   **Choix des métriques :** Différentes métriques peuvent suggérer des résultats différents, et il n'y a pas de métrique unique qui fonctionne toujours le mieux.

### 3. Modèle d'apprentissage par renforcement - Exemple : Jeu vidéo ou navigation de robot

#### Stratégie d'évaluation

Pour un modèle d'apprentissage par renforcement (par exemple, un robot naviguant dans un labyrinthe) :

1.  **Récompense cumulative :** La métrique la plus fondamentale. Mesurer la somme des récompenses obtenues par l'agent sur un épisode (une tentative complète) ou sur une période donnée. L'objectif est de maximiser cette récompense.
2.  **Taux de succès :** Si la tâche a un objectif clair (par exemple, atteindre la sortie du labyrinthe), mesurer la fréquence à laquelle l'agent atteint cet objectif avec succès.
3.  **Durée/nombre d'étapes par épisode :** Pour les tâches où l'efficacité est importante, mesurer le temps ou le nombre d'actions nécessaires pour atteindre l'objectif. Un chemin plus court ou un temps plus rapide indique une meilleure performance.
4.  **Convergence de la politique/fonction de valeur :** Surveiller l'évolution de la fonction de valeur (Q-value) ou de la politique de l'agent au fil du temps. Une convergence indique que l'agent a appris une stratégie stable.
5.  **Équilibre exploration-exploitation :** Observer si l'agent explore suffisamment l'environnement au début pour trouver des stratégies optimales, puis exploite ces stratégies pour maximiser les récompenses. Un déséquilibre peut conduire à un apprentissage sous-optimal.
6.  **Robustesse :** Tester l'agent dans des environnements légèrement modifiés ou avec des perturbations pour évaluer sa capacité à s'adapter.

#### Difficultés et limites

*   **Définition de la fonction de récompense :** Concevoir une fonction de récompense adéquate est souvent le plus grand défi. Une récompense mal définie peut conduire à des comportements indésirables ou à un apprentissage lent.
*   **Exploration-Exploitation :** Trouver le bon équilibre est difficile. Trop d'exploration ralentit l'apprentissage, trop peu peut empêcher de trouver la solution optimale.
*   **Complexité de l'environnement :** Les environnements complexes avec de nombreux états et actions peuvent rendre l'apprentissage très long et nécessiter d'énormes ressources computationnelles.
*   **Reproducibilité :** La nature stochastique de l'exploration et de l'environnement peut rendre la reproduction exacte des résultats difficile.
*   **Transfert d'apprentissage :** Les politiques apprises dans un environnement peuvent ne pas se transférer facilement à un autre, même légèrement différent.
