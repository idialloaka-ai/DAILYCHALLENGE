## Exercise 1: Open Source Levels Reflection
**Use:** Fully Open, Weights Released, Architecture Only; model components; fine-tuning/domain adaptation; healthcare compliance.

**Step-by-step**
1) Gather definitions for the three openness levels.  
2) Identify key characteristics (what is open; what you can/cannot do).  
3) Compare side by side in a table (what's open vs. impact).  
4) Write a 3?5 sentence comparative paragraph.  
5) Answer the healthcare prompt (1?2 sentences) about retraining on clinical data.

**Deliverables:** paragraph + healthcare answer.

```
import pandas as pd

# Completed definitions for open source levels
open_source_levels = {
    "Fully Open": {
        "definition": "Models where the training code, data (often curated), weights, and architecture are all released under a permissive license.",
        "what_is_open": ["Training data/scripts", "Model weights"],
        "what_you_can_do": ["Audit data bias", "Full retraining"],
        "what_you_cannot_do": ["Claim proprietary ownership of original code"],
    },
    "Weights Released": {
        "definition": "Models where the final trained weights and architecture are public, but the specific training data or exact recipe remains private.",
        "what_is_open": ["Pretrained weights", "Inference code"],
        "what_you_can_do": ["Fine-tune on private data", "Self-host locally"],
        "what_you_cannot_do": ["Reproduce exact training from scratch"],
    },
    "Architecture Only": {
        "definition": "Only the mathematical structure and layer configuration are shared, usually via a research paper, without weights.",
        "what_is_open": ["Structural design", "Paper/Methodology"],
        "what_you_can_do": ["Implement the logic", "Learn design patterns"],
        "what_you_cannot_do": ["Use the model without massive compute to train weights"],
    },
}
display(pd.DataFrame(open_source_levels).T)
```

```
comparison_table = """| Openness level | What's open? | Impact on retraining/modifying |
| --- | --- | --- |
| Fully Open | Code, weights, and data | High: Allows for complete audits and domain-specific retraining from scratch. |
| Weights Released | Model weights and architecture | Medium: Perfect for PEFT/LoRA fine-tuning on custom datasets. |
| Architecture Only | Design patterns and papers | Low: Requires significant resources to make the model functional. |"""
print(comparison_table)
```

```
comparative_paragraph = """Open-source LLMs vary significantly in their degree of transparency, ranging from fully open projects to those that only release model weights. Fully open models provide the most control, allowing developers to inspect training data for bias, while 'Weights Released' models are the industry standard for practical fine-tuning. Architecture-only releases serve primarily for academic progress but offer little immediate utility for businesses. Choosing between them involves balancing the need for deep customization against the cost of training from zero."""
print(comparative_paragraph)
```

```
healthcare_prompt_answer = """For a healthcare assistant requiring clinical retraining, the 'Weights Released' level is essential at a minimum. It allows for domain-specific fine-tuning (like LoRA) on sensitive medical data while ensuring the model stays within a secure, private infrastructure for compliance."""
print(healthcare_prompt_answer)
```
