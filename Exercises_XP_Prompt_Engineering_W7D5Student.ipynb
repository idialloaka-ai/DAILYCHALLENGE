## Exercise 1: Rewrite and Optimize a Vague Prompt
**Scenario**: Manager asks for a LinkedIn post about FlowNest with the vague prompt `"Write something about productivity tips."`

### Tasks
- Identify 3 critical issues with the prompt.
- Rewrite it using: role prompting, clear instructions, and format/tone/length constraints.
- Add a final line to make it reusable for future social posts.

```python
critical_issues = [
    'Absence de rôle défini et de public cible (LinkedIn nécessite un ton spécifique).',
    'Manque de contraintes de format (longueur, structure en puces, emojis).',
    'Objectif vague : promotion de FlowNest non mentionnée, ce qui réduit la conversion.'
]
critical_issues
```

```python
rewritten_prompt = """Tu es un expert en marketing digital spécialisé dans la productivité.
Écris un post LinkedIn engageant pour promouvoir notre nouvelle application 'FlowNest'.
Instructions : Donne 3 astuces concrètes de productivité que l'on peut appliquer avec FlowNest.
Contraintes : Utilise un ton professionnel mais enthousiaste, des puces pour la lisibilité, et ne dépasse pas 200 mots.
Réutilisabilité : Utilise ce format chaque fois que nous faisons la promotion d'outils de productivité sur les réseaux sociaux."""
rewritten_prompt
```

## Exercise 6: Hallucination Spotting and Mitigation
**Scenario**: Model added an unsupported claim in a climate/marine article summary.

### Tasks
1) Prompt the model to avoid hallucinations using only provided text.
2) Rewrite with verification language.
3) List 2 mitigation strategies.
4) Name 2 domains where hallucinations are risky and why.

```python
no_hallucination_prompt = """Résume le texte suivant.
IMPORTANT : Utilise exclusivement les informations présentes dans le texte fourni. Ne fais appel à aucune connaissance externe."""
no_hallucination_prompt
```

```python
verification_prompt = """Résume l'article. N’incluez que les affirmations explicitement mentionnées.
En cas de doute ou si une information n’est pas présente, indique explicitement que l’information n’est pas disponible dans le texte."""
verification_prompt
```

```python
mitigation_strategies = [
    'Utilisation de consignes de "grounding" (ancrage) strictes.',
    'Ajout d\'une instruction de aveu d\'ignorance ("si tu ne sais pas, dis-le").'
]
risky_domains = {
    'domain_1': {'name': 'Médical', 'why': 'Une dose erronée ou un diagnostic faux peut mettre des vies en danger.'},
    'domain_2': {'name': 'Juridique', 'why': 'L\'invention de précédents judiciaires peut entraîner des sanctions graves et fausser le droit.'},
}
display(mitigation_strategies, risky_domains)
```
