### 1.1 Define LLMs

Large Language Models (LLMs) are advanced artificial intelligence models trained on vast amounts of text data. They are designed to understand, generate, and process human language. Their primary tasks include:

## 🌟 Exercise 2 · Transformer Architecture and Tokenization


### 2.1 Explain Tokenization

Tokenization is the process of breaking down a sequence of text into smaller units called "tokens." These tokens can be words, subwords, or even characters, depending on the tokenizer used. For LLMs, tokenization is a crucial first step because models don't directly understand raw text; they operate on numerical representations.

The process typically involves:
1.  **Splitting text**: The input text is split into chunks based on rules (e.g., spaces, punctuation).
2.  **Vocabulary mapping**: Each token is then mapped to a unique numerical ID from a predefined vocabulary. If a token isn't in the vocabulary, it might be split further into subword units until all parts are recognizable.

This conversion allows the model to process text efficiently as numerical data.


```python
text = "Bonjour, le monde des grands modèles de langage est fascinant !"  # TODO: provide a sentence of your choice
if text is None:
    raise ValueError("Define the variable `text` with a short sentence.")

tokens = tokenizer.tokenize(text)         # tokenize the input
token_ids = tokenizer.convert_tokens_to_ids(tokens)  # convert to ids

print(f"Original Text: {text}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")

x_label = "Tokens"  # label for the x-axis (e.g., "Tokens")
y_label = "Token IDs"  # label for the y-axis (e.g., "Token IDs")
title = "Tokenization Process Visualization"    # chart title

if None in (x_label, y_label, title):
    raise ValueError("Set x_label, y_label, and title before plotting.")

plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color="skyblue")
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.xticks(rotation=45)
plt.show()
```

### 3.2 Explain special prefixes

In GPT-style vocabularies, the special prefix `Ġ` (a Unicode character representing a space) indicates that the token it precedes starts with a space. This is common in Byte Pair Encoding (BPE) or WordPiece tokenizers, which are often used in models like GPT-2. The purpose is to preserve information about word boundaries while allowing for subword tokenization.

For example:
*   `['ĠHello']` means the word "Hello" started with a space.
*   `['world']` means "world" is part of a larger word or follows a non-space character.

This approach helps in reconstructing the original text accurately and handles unknown words by breaking them down into known subword units, while still maintaining context about word separation.

## 🌟 Exercise 4: Pré-entraînement vs. Mise au point

LLMs undergo two main phases of training:

### 4.1 Pré-entraînement (Pre-training)

Pre-training is the initial phase where a large language model is trained on a massive and diverse dataset of text (e.g., books, articles, web pages). During this phase, the model learns general linguistic patterns, grammar, factual knowledge, and reasoning abilities. It typically uses self-supervised learning objectives, such as predicting the next word in a sequence (causal language modeling, like GPT) or filling in masked words (masked language modeling, like BERT).

The goal of pre-training is to create a general-purpose language understanding and generation model that can be adapted to various downstream tasks.

### 4.2 Réglage fin (Fine-tuning)

Fine-tuning is the subsequent phase where the pre-trained model is adapted to a specific downstream task (e.g., sentiment analysis, question answering, summarization) or a particular domain (e.g., legal text, medical records). This involves training the model on a smaller, task-specific dataset, often with labeled examples.

During fine-tuning, the model's weights are slightly adjusted to optimize its performance on the target task. Because the model has already learned extensive linguistic knowledge during pre-training, it requires much less data and training time to achieve high performance on new tasks compared to training a model from scratch.

In essence:
*   **Pre-training:** General knowledge acquisition from vast, unlabeled data.
*   **Fine-tuning:** Task-specific adaptation from smaller, labeled data.
