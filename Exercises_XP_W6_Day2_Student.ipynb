## 🌟 Exercise 1: Traditional vs. Modern NLP: Comparative Analysis
1) Complete the table:

| Aspect | Traditional NLP | Modern NLP |
|---|---|---|
| Feature Engineering | Manual (TF-IDF, POS tagging, hand-crafted rules) | Automatic (Learned through neural network layers) |
| Word Representations | Static/Discrete (One-hot, Word2Vec, GloVe) | Contextualized (Dynamic embeddings based on surrounding words) |

2) Discuss: How did the shift to modern NLP impact scalability and efficiency?

The shift to modern NLP revolutionized scalability by moving away from labor-intensive manual feature engineering toward representation learning, where models autonomously extract features from vast datasets. This allowed for the utilization of massive, unlabelled datasets through self-supervised learning, drastically reducing the dependency on expensive human-labeled data. Architecturally, the move from sequential processing (like RNNs) to the Transformer architecture enabled hardware parallelism, allowing models to be trained efficiently on distributed GPU clusters. Furthermore, transfer learning meant that a single pre-trained model could be adapted to dozens of downstream tasks with minimal additional compute. This efficiency has made it possible to deploy highly capable language interfaces at a global scale.

## 🌟 Exercise 2: LLM Architecture and Application Scenarios
For each, describe (a) core architectural differences, (b) a real application, (c) why it fits.

### BERT
- **Architecture**: Bidirectional encoder-only; Masked Language Modeling (MLM). It looks at context from both left and right simultaneously.
- **Application**: Sentiment Analysis or Named Entity Recognition (NER).
- **Why**: Because BERT understands the relationship between words in a full sentence context simultaneously, it is excellent at classification tasks where the 'gist' of the whole input matters more than generating new text.

### GPT
- **Architecture**: Decoder-only; autoregressive causal LM. It predicts the next token based strictly on previous tokens (left-to-right).
- **Application**: Creative writing assistants or Code generation.
- **Why**: Its causal nature makes it highly optimized for sequence generation, allowing it to produce coherent, long-form text that flows naturally from a prompt.

### T5
- **Architecture**: Encoder-decoder; text-to-text framework. It converts every NLP task into a string-to-string format.
- **Application**: Abstractive Summarization or Language Translation.
- **Why**: The encoder builds a rich representation of the source text, while the decoder generates the target text, making it perfect for tasks where the input and output have different structures but related meanings.

## 🌟 Exercise 3: Benefits and Ethics of Pre-training
- **Benefits**:
  1) **Improved generalization**: Models learn universal language patterns that apply to many domains.
  2) **Less labeled data**: You only need a few hundred examples for fine-tuning instead of millions.
  3) **Faster fine-tuning**: Starting from a pre-trained state reaches high accuracy in fewer epochs.
  4) **Transfer learning**: Knowledge from a general task (predicting words) transfers to specific tasks (legal analysis).
  5) **Robustness**: Exposure to diverse internet data makes models more resilient to variations in syntax and vocabulary.

- **Ethical concerns**: Models can inherit social biases (gender/racial) from training data, generate hallucinations (misinformation), be used for phishing/deepfakes, or leak private information present in the training set.
- **Mitigations**: Implementing rigorous data curation to remove toxic content, using RLHF (Reinforcement Learning from Human Feedback) to align model behavior, applying differential privacy, and performing regular red-teaming/safety audits.

## 🌟 Exercise 4: Transformer Architecture Deep Dive
### Self-Attention & Multi-Head Attention
- **Q/K/V Flow**: Queries (what I'm looking for), Keys (what I offer), and Values (what I contain) interact via dot-product. Softmax creates a weight distribution (importance) used to compute a weighted sum of Values.
- **Multi-Head**: Different heads project the same input into different subspaces. One head might focus on grammar (subject-verb), while another focuses on semantic meaning (contextual synonyms).
- **Example**: "The bank of the river was flooded, so the man didn't go to the bank to withdraw money."
  - **Head 1**: Links the first "bank" to "river" (geographical relation).
  - **Head 2**: Links the second "bank" to "withdraw" (financial relation).

### Pre-training Objectives
- **MLM vs CLM**: MLM hides tokens in the middle (predict [MASK]), whereas CLM predicts the very next token. MLM is better for understanding; CLM is better for generating.
- **Selection**: Prefer MLM for extraction/classification; prefer CLM for chatbots/stories.
- **NSP**: Next Sentence Prediction helped BERT understand relationships between sentences. Modern models avoid it because the task is often too easy and doesn't improve performance as much as training on longer sequences with MLM alone.

### Transformer Model Selection
- **Sentiment on reviews**: **Encoder-only (BERT)**. Justification: It provides a deep bidirectional representation of the review, perfect for classifying the overall sentiment.
- **Conversational chatbot**: **Decoder-only (GPT)**. Justification: Chat requires fluent, step-by-step token generation based on the history.
- **Technical translation**: **Encoder-Decoder (T5/BART)**. Justification: The encoder masters the source technical terminology, and the decoder maps it accurately to the target language syntax.

### Positional Encoding
- **Purpose**: Transformers process all words in parallel and have no inherent sense of order. Positional encodings add a unique signal to each word embedding to denote its place in the sequence.
- **Failure**: Without positions, "The dog bit the man" and "The man bit the dog" would have identical representations to the model.

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator
### 1) Temperature Scenarios
- **T=0.2**: Highly deterministic. The model will almost always pick the most likely word (high confidence, low variety).
- **T=1.5**: High randomness. The model explores unlikely tokens, often leading to creative but potentially nonsensical or 'hallucinated' text.
- **T=1.0**: Standard behavior as seen during training; a balance of logical flow and natural variety.

### 2) Application Design
- **Bedtime stories**: Use **T=0.8 to 1.2**. This allows for creative plot twists and imaginative vocabulary while maintaining a coherent story arc.
- **Financial report summaries**: Use **T=0.1 or 0.0**. Accuracy is paramount; the model should only output the most probable, factual summary without 'creative' embellishments.

### 3) Temperature & Bias
- Temperature can amplify bias. At high temperatures (T>1.0), the model might drift into toxic or biased stereotypes that were weakly present in the training data. Conversely, very low temperature (T=0) might consistently output a single biased viewpoint if that was the dominant pattern in the training set. Example: In job descriptions, a low T might always associate 'Doctor' with 'He', whereas a higher T might occasionally vary the pronoun but introduce other inaccuracies.

### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"
prompt = "The future of humanity and artificial intelligence involves"

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    # Use a fixed seed for reproducibility in a demo
    torch.manual_seed(42)
    out = pipe(prompt, max_new_tokens=40, temperature=T, do_sample=True)
    print(f"\n--- Temperature: {T} ---")
    print(out[0]["generated_text"])
```
