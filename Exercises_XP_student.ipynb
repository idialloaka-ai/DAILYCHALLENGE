**Answers:**

- **What is the key difference between traditional machine learning and deep learning?**
  The key difference is in feature extraction. Traditional machine learning requires manual feature engineering, where humans design and extract relevant features from the data. Deep learning, on the other hand, automatically learns hierarchical features directly from raw data through its multiple layers, eliminating the need for manual feature engineering.

- **How do artificial neural networks mimic the human brain?**
  Artificial neural networks (ANNs) mimic the human brain by using interconnected nodes (neurons) organized in layers. Each connection has a weight, similar to synapses in the brain, and neurons activate based on input signals, processing information in a distributed and parallel manner, much like biological neurons.

- **Why does deep learning perform better on large datasets compared to traditional machine learning?**
  Deep learning models have a high capacity to learn complex patterns and representations from data. With larger datasets, deep learning models can learn more robust and generalized features, leading to better performance and reduced overfitting, whereas traditional machine learning algorithms often struggle with the sheer volume and complexity of large datasets without extensive feature engineering.

- **What are some challenges of deep learning, and how can they be addressed?**
  Challenges include the need for vast amounts of data, high computational resources, and the 'black-box' nature of models (lack of interpretability). These can be addressed by techniques like data augmentation, transfer learning (for limited data), using powerful GPUs/TPUs, and developing explainable AI (XAI) methods for interpretability.

- **What is feature engineering, and why is it not needed in deep learning?**
  Feature engineering is the process of manually creating new input features from existing raw data to improve the performance of machine learning models. It is not strictly needed in deep learning because deep neural networks, with their multiple layers, can automatically learn and extract relevant features from the raw input data itself, building hierarchical representations.

- **What role do hidden layers play in a deep learning model?**
  Hidden layers are intermediate layers between the input and output layers that perform complex transformations on the input data. They learn to identify and extract abstract, hierarchical features from the raw input, allowing the network to recognize intricate patterns and relationships that are not immediately obvious in the raw data.

- **In an artificial neural network, what is the function of an activation function?**
  The function of an activation function is to introduce non-linearity into the network. Without activation functions, a neural network would only be able to learn linear relationships, regardless of how many layers it has. Activation functions allow the network to learn and model complex, non-linear patterns in the data.

## 🌟 Exercise 3: Building a Simple Neural Network with TensorFlow or Keras

**As stated in the instructions**  
Build a classifier for MNIST with:
- One input layer
- One hidden layer with 128 neurons and ReLU activation
- One output layer with 10 neurons and softmax activation

Steps:
1. Import TensorFlow and needed modules from `tensorflow.keras`
2. Load the MNIST dataset
3. Normalize data by dividing by 255.0
4. One-hot encode the labels
5. Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
6. Compile with optimizer `adam`, loss `categorical_crossentropy`, metrics `accuracy`
7. Train and evaluate on the test set

**Guidance**  
You will execute prefilled boilerplate for imports and dataset loading. You will complete the one-hot encoding and the model specification, compile, train, and evaluate.

**Learning point**  
Normalization stabilizes optimization. One-hot encoding matches softmax outputs to target vectors. ReLU accelerates training by avoiding saturation, while softmax maps logits to a probability simplex.

![image.png](https://github.com/user-attachments/assets/61737b80-69cb-41a1-aea5-76092cd2a373)


```python
# PREFILLED: just execute
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist

print("TensorFlow:", tf.__version__)

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Flatten will be added in the model; keep images as 28x28 for now
print("Train shape:", x_train.shape, "Test shape:", x_test.shape)
```

```python
# To-Do: one-hot encode labels y_train and y_test using to_categorical
# Hints: use to_categorical
y_train_oh = to_categorical(y_train, num_classes=10)
y_test_oh  = to_categorical(y_test, num_classes=10)

print("Labels one-hot:", y_train_oh.shape, y_test_oh.shape)
```

```python
# To-Do:  Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)), # Input layer needs input_shape
    layers.Dense(128, activation='relu'), # Hidden layer with 128 neurons and ReLU
    layers.Dense(10, activation='softmax'), # Output layer with 10 neurons and softmax
])

model.summary()
```

```python
# To-Do: compile the model with adam, categorical_crossentropy, and accuracy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# To-Do: train then evaluate
history = model.fit(x_train, y_train_oh, epochs=5, batch_size=32, validation_split=0.2)
test_loss, test_acc = model.evaluate(x_test, y_test_oh, verbose=0)
print({"test_loss": float(test_loss), "test_acc": float(test_acc)})
```

## 🌟 Exercise 4: Forward Propagation Calculation

**As stated in the instructions**  
Predict house prices given:
- Inputs: x₁ = 2000 square feet, x₂ = 3 bedrooms
- Weights: w₁ = 0.5, w₂ = 0.7
- Bias: b = 50,000
- Activation: ReLU

Compute z before activation, apply ReLU, and interpret the prediction.

**Guidance**  
Write scalar computations. Then reason about the role of the bias and ReLU on the output scale.

**Learning point**  
ReLU passes positive linear combinations unchanged and zeros out negatives. Bias sets the baseline prediction independent of inputs.
![](./figs/1.gif)

```python
# To-Do: compute forward propagation for the given numbers
x1, x2 = 2000.0, 3.0
w1, w2 = 0.5, 0.7
b = 50000.0

# To-Do: compute z = x1*w1 + x2*w2 + b
z = x1*w1 + x2*w2 + b

# To-Do: apply ReLU
y = max(0, z) #Hint: ReLU acts like max functions and leaves only positive outputs

print({"z": z, "prediction": y})
```

**To-Do:** Interpret the result as a house price. Explain the influence of input scaling and bias in full sentences.

**Interpretation:**

The calculated `z` value (before activation) is `51002.1`. After applying the ReLU activation function, the final prediction `y` is also `51002.1` because `z` is positive. Therefore, the predicted house price is **$51,002.1**.

**Influence of Input Scaling and Bias:**

*   **Input Scaling:** The weights (`w1=0.5`, `w2=0.7`) effectively scale the input features (`x1=2000` sq ft, `x2=3` bedrooms). This scaling determines the relative importance or contribution of each feature to the final prediction. For example, `x1` (square footage) has a much larger direct impact on `z` due to its larger value, even with a smaller weight, compared to `x2` (number of bedrooms).

*   **Bias:** The bias term (`b=50000`) acts as a baseline or intercept. It ensures that even if all input features were zero, there would still be a non-zero prediction. In this context, it sets a fundamental price floor or a starting point for the house value, independent of the size or number of bedrooms, capturing inherent market value. Without the bias, the predicted price would be `2000 * 0.5 + 3 * 0.7 = 1002.1`, which is much lower and likely unrealistic, highlighting the bias's role in anchoring the prediction to a meaningful range.

## 🌟 Exercise 5 (optional): Implementing Forward and Backward Propagation in Python

**As stated in the enoncé**  
You will code a simple neural network that performs forward propagation and backpropagation for a regression problem predicting exam scores from study hours.

**Guidance**  
Run the prefilled code. Then modify the learning rate or initial weights and observe changes in the loss and updates.
Provide a short written explanation.

**Learning point**  
Gradient descent updates parameters in the direction of steepest loss reduction.
The learning rate controls step size. Too small slows learning. Too large risks divergence.


```python
# PREFILLED: just execute
import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)
```

**To-Do:** Explain in full sentences why updating weights using gradient descent reduces the error. Then modify the learning rate or initial weights and describe how this affects learning.

**Explanation of Gradient Descent:**

Updating weights using gradient descent reduces the error by iteratively adjusting the model's parameters (weights and bias) in the direction opposite to the gradient of the loss function. The gradient points towards the steepest increase in the loss. By moving in the opposite direction, gradient descent effectively takes small steps down the "error landscape," gradually minimizing the loss function and improving the model's accuracy. Each update brings the predicted output closer to the true output.

**Impact of Modifying Learning Rate or Initial Weights:**

If we were to **modify the learning rate**:
*   **Increasing the learning rate (e.g., from 0.01 to 0.1):** The `w_new` and `b_new` values would change more dramatically in a single step. While this might lead to faster convergence if the learning rate is well-chosen, too high a learning rate could cause the optimization to overshoot the minimum loss, oscillate, or even diverge, preventing the model from converging to an optimal solution.
*   **Decreasing the learning rate (e.g., from 0.01 to 0.001):** The updates to `w_new` and `b_new` would be smaller. This would result in slower convergence, requiring more iterations (epochs) to reach the minimum loss. However, a smaller learning rate can help in finding a more precise minimum and avoid overshooting, especially in complex loss landscapes.

If we were to **modify the initial weights (w)**:
*   **Changing initial weights (e.g., `w = np.array([0.1, 0.9])` instead of `[0.6, 0.3]`):** The initial `y_pred` and `loss` would be different. This would change the starting point of the optimization process on the error landscape. Different starting points can lead to the model converging to different local minima in more complex models or taking a different path to the same global minimum. In this simple linear case, it primarily affects the initial magnitude and direction of `grad_w` and `grad_b`, thus altering the first update step.
