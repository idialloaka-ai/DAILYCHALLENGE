 Qu'est-ce que l'analyse de données ?
L'analyse de données est un processus systématique qui consiste à inspecter, nettoyer, transformer et modéliser des données brutes. Son but ultime est d'en extraire des informations utiles, de formuler des conclusions claires et d'éclairer la prise de décision. Elle transforme des faits bruts non organisés en connaissances stratégiques exploitables.


 Pourquoi est-elle importante dans les contextes modernes ?
À l'ère du Big Data, les organisations sont submergées d'informations. Sans analyse, ces données sont inutiles. L'analyse moderne permet de :

Réduire l'incertitude : Remplacer les intuitions par des preuves basées sur les faits.

Optimiser les processus : Repérer les inefficacités en temps réel pour économiser des ressources.

Personnaliser l'expérience : Comprendre précisément les comportements des utilisateurs pour anticiper leurs besoins.

Trois domaines d'application actuels
La Santé (Médecine de précision) : Analyse des données cliniques et génétiques des patients pour prédire l'apparition de maladies et personnaliser les traitements médicaux.

Le E-commerce et le Commerce de détail : Utilisation des historiques d'achat et de navigation pour alimenter les algorithmes de recommandation (comme sur Netflix ou Amazon) et optimiser la gestion des stocks.

La Finance (Détection de la fraude) : Surveillance automatisée des transactions bancaires en temps réel. Les modèles identifient les anomalies de comportement (ex: un achat inhabituel à l'étranger) pour bloquer instantanément les cartes piratées.

Raisonnement : Elles contiennent des chaînes de caractères (type object). Elles représentent des catégories non mesurables (comme le type d'emploi ou le statut de validation de la carte + / -).

Exercice 6 : Identification des types de données (Structuré vs Non Structuré)
Rapports financiers dans Excel : Données Structurées. Les données sont organisées de manière rigide en lignes, colonnes et cellules identifiables.

Photographies sur les réseaux sociaux : Données Non Structurées. Il s'agit de matrices de pixels brutes, sans structure tabulaire native.

Articles de presse sur un site web : Données Non Structurées. C'est du texte libre narratif de longueurs variables.

Données d'inventaire dans une base relationnelle : Données Structurées. Suivent un schéma SQL strict avec des clés primaires et des types définis.

Enregistrements d'entretiens (étude de marché) : Données Non Structurées. Fichiers audio ou transcriptions de discussions libres.

```python
# Exercice 9: Exporter un DataFrame au format Excel et JSON


import pandas as pd

# 1. Création d'un dataframe simple
donnees = {
    'Nom': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Ville': ['Paris', 'Lyon', 'Marseille']
}
df_simple = pd.DataFrame(donnees)

# 2. Exportation au format Excel
df_simple.to_excel('mon_fichier.xlsx', index=False)
print("✅ Fichier Excel 'mon_fichier.xlsx' exporté avec succès.")

# 3. Exportation au format JSON
df_simple.to_json('mon_fichier.json', orient='records', indent=4)
print("✅ Fichier JSON 'mon_fichier.json' exporté avec succès.")
```
