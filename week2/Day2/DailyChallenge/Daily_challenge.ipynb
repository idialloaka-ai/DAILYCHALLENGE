```python
# Daily challenge




# Importation des bibliothèques
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder


# Remplace le nom du fichier par le tien
df = pd.read_csv("datascience_salaries.csv")

# Afficher les premières lignes
print("First 5 rows of the dataset:\n")
print(df.head())




scaler = MinMaxScaler()

df["salary_normalized"] = scaler.fit_transform(df[["salary"]])

print("\nNormalized Salary Column:\n")
print(df[["salary", "salary_normalized"]].head())




# Copier le dataset
pca_df = df.copy()

# Convertir les colonnes texte en nombres
label_encoder = LabelEncoder()

for column in pca_df.select_dtypes(include=["object"]).columns:
    pca_df[column] = label_encoder.fit_transform(pca_df[column])

# Retirer les valeurs manquantes si nécessaire
pca_df = pca_df.dropna()

# Appliquer PCA
pca = PCA(n_components=2)

principal_components = pca.fit_transform(pca_df)

# Créer un DataFrame avec les nouvelles dimensions
pca_result = pd.DataFrame(
    data=principal_components,
    columns=["Principal_Component_1", "Principal_Component_2"]
)

print("\nPCA Result:\n")
print(pca_result.head())



salary_stats = df.groupby("experience_level")["salary"].agg(
    ["mean", "median"]
)

print("\nAverage and Median Salary by Experience Level:\n")
print(salary_stats)



print("\nExplanation:")
print("- Min-Max normalization scales salary values between 0 and 1.")
print("- PCA reduces the number of features while keeping important information.")
print("- Aggregation helps analyze salary trends by experience level.")
```
