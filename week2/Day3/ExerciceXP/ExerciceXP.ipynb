```python
# Exe"rcice 2

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean     = np.mean(data)
median   = np.median(data)
variance = stats.tvar(data)    # variance d'échantillon (ddof=1)
std_dev  = stats.tstd(data)     # écart-type d'échantillon

print(f"Moyenne   : {mean}")
print(f"Médiane   : {median}")
print(f"Variance  : {variance:.2f}")
print(f"Écart-type: {std_dev:.2f}")
```

```python
# Exercice 4

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)
print(f"T-statistic : {t_stat:.4f}")
print(f"P-value     : {p_value:.6f}")
```

```python
# Exercice 5 


house_sizes  = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

slope, intercept, r, p, se = stats.linregress(house_sizes, house_prices)

pred_90 = slope * 90 + intercept
print(f"Pente (slope)   : {slope:.2f} €/m²")
print(f"Intercept       : {intercept:.2f}")
print(f"R²              : {r**2:.4f}")
print(f"Prix pour 90 m² : {pred_90:.0f} €")
```

```python
# Exercice 6

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_val, p_val = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print(f"F-value : {f_val:.4f}")
print(f"P-value : {p_val:.6f}")
```
