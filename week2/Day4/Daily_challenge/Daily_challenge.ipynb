```python
# 1. Préparation et Nettoyage des Données
# Chargement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets

df = pd.read_csv("superstore_dataset.csv")
# Vérification initiale
print(df.shape)
print(df.info())
print(df.isnull().sum())
# Gestion des doublons
print("Doublons :", df.duplicated().sum())

df = df.drop_duplicates()
# Gestion des valeurs manquantes
df['Postal Code'] = df['Postal Code'].fillna(0)
# Conversion des dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
# Création de variables
df['Profit Margin'] = (df['Profit']/df['Sales'])*100

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month-Year'] = df['Order Date'].dt.to_period('M')
# 2. Analyse Exploratoire
# Chiffres globaux
print("Ventes totales :", df['Sales'].sum())
print("Profit total :", df['Profit'].sum())
print("Marge moyenne :", df['Profit Margin'].mean())

# 3. Analyse Temporelle
# Ventes mensuelles
monthly_sales = df.groupby('Order Month-Year')['Sales'].sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index.astype(str),
         monthly_sales.values)

plt.xticks(rotation=90)
plt.title("Évolution des ventes mensuelles")
plt.show()
"""Observations
Forte croissance progressive.
Pics récurrents en novembre et décembre.
Saisonnalité liée aux fêtes."""
# 4. Analyse Géographique
# Top États par ventes
state_sales = df.groupby('State')['Sales'].sum()

top_states = state_sales.sort_values().tail(10)

plt.figure(figsize=(10,6))
top_states.plot(kind='barh')
plt.title("Top 10 États par ventes")
plt.show()

""" interprétation
Les ventes sont fortement concentrées sur quelques grands marchés."""

# 5. Produits les Plus Rentables
top_products = (
    df.groupby('Product Name')['Profit']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,8))

sns.barplot(
    x=top_products.values,
    y=top_products.index
)

plt.title("Top 10 Produits les Plus Rentables")
plt.show()
"""Conclusion

Les produits technologiques dominent généralement le classement :

Copieurs
Imprimantes
Machines professionnelles"""
# 6. Analyse Remise vs Profit
plt.figure(figsize=(12,7))

sns.scatterplot(
    data=df,
    x='Discount',
    y='Profit',
    hue='Category'
)

sns.regplot(
    data=df,
    x='Discount',
    y='Profit',
    scatter=False,
    color='red'
)

plt.axhline(0,color='black')
plt.show()

# Analyse des remises >20 %
high_discount = df[df['Discount'] > 0.20]

loss_rate = (
    high_discount['Profit'] < 0
).mean()*100

print(loss_rate)
"""Observation

Environ :

60 % à 70 %

des ventes avec plus de 20 % de remise deviennent déficitaires."""

# 7. Analyse par Catégorie
category_sales = (
    df.groupby('Category')
      .agg({
          'Sales':'sum',
          'Profit':'sum'
      })
)

print(category_sales)

```
