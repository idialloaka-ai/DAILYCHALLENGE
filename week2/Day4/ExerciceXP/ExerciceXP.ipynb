Exercice 1 : Comprendre la visualisation des données


La visualisation des données est une étape essentielle de l'analyse de données car elle permet de transformer des données brutes en informations compréhensibles. Grâce aux graphiques, il devient plus facile d'identifier des tendances, des anomalies, des corrélations et des modèles cachés qui seraient difficiles à détecter dans un tableau de données.

Les visualisations facilitent également la communication des résultats aux décideurs et aux parties prenantes en présentant les informations de manière claire et intuitive.

Importance de la visualisation des données
Simplifie l'interprétation des données.
Met en évidence les tendances et les motifs.
Permet de détecter rapidement les anomalies.
Facilite la prise de décision.
Améliore la communication des résultats.
Utilité d'un graphique linéaire

Un graphique linéaire est utilisé pour représenter l'évolution d'une variable au fil du temps.

Il permet :

d'observer les tendances générales ;
d'identifier les hausses et les baisses ;
de comparer plusieurs séries temporelles ;
d'analyser les variations sur une période donnée.

Les graphiques linéaires sont particulièrement adaptés aux données chronologiques telles que les températures, les ventes mensuelles ou les cours boursiers.

```python
import matplotlib.pyplot as plt

jours = ["Lun", "Mar", "Mer", "Jeu", "Ven", "Sam", "Dim"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.figure(figsize=(8,5))

plt.plot(
    jours,
    temperatures,
    marker='o',
    linewidth=2,
    color='blue'
)

plt.title("Variation de la température sur une semaine")
plt.xlabel("Jour")
plt.ylabel("Température (°F)")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

mois = ["Jan", "Fév", "Mar", "Avr", "Mai"]
ventes = [5000, 5500, 6200, 7000, 7500]

plt.figure(figsize=(8,5))

plt.bar(
    mois,
    ventes,
    color='skyblue'
)

plt.title("Ventes mensuelles du magasin")
plt.xlabel("Mois")
plt.ylabel("Montant des ventes ($)")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv("Student Mental health.csv")

plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x='What is your CGPA?',
    color='purple'
)

plt.title("Distribution des catégories de CGPA")
plt.xlabel("CGPA")
plt.ylabel("Nombre d'étudiants")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Student Mental health.csv")

# Conversion Oui/Non en 1/0
df['Panic_Attack_Numeric'] = (
    df['Do you have Panic attack?']
    .map({'Yes':1, 'No':0})
)

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_Attack_Numeric',
    color='red',
    s=80,
    alpha=0.7
)

plt.title("Relation entre l'âge et les crises de panique")
plt.xlabel("Âge")
plt.ylabel("Crise de panique (1=Oui, 0=Non)")

plt.yticks([0,1], ['Non', 'Oui'])

plt.show()
```
