```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ── 1. PREPARATION DES DONNEES ──────────────────────────────────────────────

np.random.seed(42)

temperatures = np.random.uniform(low=-5, high=35, size=(10, 12))

cities = [f"City {i}" for i in range(1, 11)]
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

df = pd.DataFrame(temperatures, index=cities, columns=months)

print("Apercu du DataFrame :")
print(df.round(2))

# ── 2. ANALYSE DES DONNEES ───────────────────────────────────────────────────

annual_mean = df.mean(axis=1)

hottest_city  = annual_mean.idxmax()
coldest_city  = annual_mean.idxmin()

print("\nTemperature moyenne annuelle par ville :")
print(annual_mean.round(2))
print(f"\nVille la plus chaude : {hottest_city} ({annual_mean[hottest_city]:.2f} C)")
print(f"Ville la plus froide : {coldest_city} ({annual_mean[coldest_city]:.2f} C)")

# ── 3. VISUALISATION ─────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(14, 7))

colors = plt.cm.winter(np.linspace(0, 1, len(cities)))

for i, city in enumerate(cities):
    linestyle = "--" if city in [hottest_city, coldest_city] else "-"
    linewidth = 2.5 if city in [hottest_city, coldest_city] else 1.2
    ax.plot(months, df.loc[city], label=city,
            color=colors[i], linestyle=linestyle, linewidth=linewidth)

ax.set_title("Monthly Temperature Trends for Each City", fontsize=16, pad=15)
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Temperature (C)", fontsize=12)
ax.legend(loc="lower left", fontsize=9)
ax.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.savefig("temperature_trends.png", dpi=150)
plt.show()

# ── 4. RAPPORT DE SYNTHESE ───────────────────────────────────────────────────

print("\n========== RAPPORT DE SYNTHESE ==========")
print(f"Jeu de donnees : 10 villes x 12 mois, temperatures entre -5 et 35 C")
print(f"\nVille la plus chaude : {hottest_city} — moyenne annuelle {annual_mean[hottest_city]:.2f} C")
print(f"Ville la plus froide : {coldest_city} — moyenne annuelle {annual_mean[coldest_city]:.2f} C")
print(f"\nMoyenne globale toutes villes confondues : {annual_mean.mean():.2f} C")
print(f"Mois le plus chaud en moyenne : {df.mean(axis=0).idxmax()}")
print(f"Mois le plus froid en moyenne  : {df.mean(axis=0).idxmin()}")
print("\nTendance : les donnees etant aleatoires, aucune saisonnalite")
print("systematique n'est observee — ce qui serait different avec")
print("des donnees reelles ou une simulation par hemisphere.")
print("=========================================")
```
